```python
import tensorflow as tf
from tensorflow.keras import layers
from tensorflow.keras import models
from tensorflow import keras
```

```python
# 相对 mnist1 增加了层数
# Network and training.
EPOCHS = 50  # 更少
BATCH_SIZE = 128
VERBOSE = 1
NB_CLASSES = 10  # number of outputs = number of digits
N_HIDDEN = 128
VALIDATION_SPLIT = 0.2  # how much TRAIN is reserved for VALIDATION
```

```python
(X_train, Y_train), (X_test, Y_test) = keras.datasets.mnist.load_data()

RESHAPED = 784
X_train = X_train.reshape(60000, RESHAPED)
X_test = X_test.reshape(10000, RESHAPED)

X_train = X_train.astype('float32')
X_test = X_test.astype('float32')

X_train, X_test = X_train / 255.0, X_test / 255.0
```

```python
Y_train = keras.utils.to_categorical(Y_train, NB_CLASSES)
Y_test = keras.utils.to_categorical(Y_test, NB_CLASSES)
```

```python
models = models.Sequential()
models.add(layers.Dense(N_HIDDEN,  # 输出 128
                        input_shape=(RESHAPED,),
                        name='dense_layer',
                        activation='relu'))
models.add(layers.Dense(N_HIDDEN,  # 后面不需要定义 input_shape，会自动推断
                        name='dense_layer_2',
                        activation='relu'))
models.add(layers.Dense(NB_CLASSES,  # 最后一层，输出 10，以 softmax 转换为概率
                        name='dense_layer_3',
                        activation='softmax'))
models.summary()
```

```python
models.compile(optimizer='SGD',
               loss='categorical_crossentropy',
               metrics=['accuracy'])
models.fit(X_train, Y_train,
           batch_size=BATCH_SIZE,
           epochs=EPOCHS,
           verbose=VERBOSE,
           validation_split=VALIDATION_SPLIT)
```

```python
# 97.16%, 96.57,轻微的过拟合
test_loss, test_acc = models.evaluate(X_test, Y_test)

# 96.35，还可以
```
